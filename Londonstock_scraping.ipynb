```python
url = 'http://londonstockexchange.com/exchange/prices-and-markets/stocks/prices-search/stock-prices-search.html?&page=757'
response = requests.get(url)
html = response.text.encode('utf-8')
def getkeys(html):
    soup = BeautifulSoup(html,'html5lib')
    #tmp = soup.find('tr',{'class' :'name'})
    tmp = soup.findAll('td', {'class' :'name'})
    count=0
    for tag in tmp:   
        try:
            a = tag.find('a', href=True)['href']
            keyurls.append('http://londonstockexchange.com/exchange/prices/stocks/summary/fundamentals.html?fourWayKey='+a.split('/')[6].split('.')[0])
            count+=1
        except:
            pass
    print(count)
    div = soup.find("div", {"class" : "paging"})

    hrefs = div.findAll('a', href= True)
    for href in hrefs:
        if href.find(text = True) == 'Next':
            url = 'http://londonstockexchange.com'+href['href']
            if url[-4:] == '1226':
                return
            print(url)
            response = requests.get(url)
            html = response.text.encode('utf-8')
            time.sleep(0.5)
            return getkeys(html)


getkeys(html)
```

```python
url = 'https://www.londonstockexchange.com/exchange/prices/stocks/summary/fundamentals.html?fourWayKey=GB00BCDBXK43GBGBXASX1'
response = requests.get(url)
html = response.text.encode('utf-8')
def get_columns(html):
    columns = ['Company Name']
    soup = BeautifulSoup(html,'html5lib')
    trs = soup.find_all('tr')
    for tr in trs:
        ui = []
        for td in tr:
            ui.append(str(td.string).replace('\n',''))    
        if len(ui)>5 and ui[3] != 'None' and ui[1] != '\xa0' and ui[1] != 'None':
            columns.append(ui[1].replace('\xa0',''))
    columns += ['FTSE ICB sector','FTSE ICB subsector','Company market cap, m*']
    
    return(columns)
columns = get_columns(html)
columns[44] = 'Closing date'
```

```python
df = {}
df.update({'13':[]})
df.update({'14':[]})
df.update({'15':[]})
df.update({'16':[]})
df.update({'17':[]})
```

```python
# url = 'https://www.londonstockexchange.com/exchange/prices/stocks/summary/fundamentals.html?fourWayKey=GB0000066554GBGBXSTMM'
# #url = 'https://londonstockexchange.com/exchange/prices/stocks/summary/fundamentals.html?fourWayKey=GB00B1YW4409GBGBXSET1'
# response = requests.get(url)
# html = response.text.encode('utf-8')
def chart_data(html,df):
    soup = BeautifulSoup(html,'html5lib')
    try:
        tmp = soup.find('ol', {'class' :'breadcrumb'}).find('li', {'class' :'active'})
    except:
        return
    name = str(tmp).split('>')[2].split('\n')[0]
    trs = soup.find_all('tr')
    tmp=[[name],[name],[name],[name],[name]]
    try:
        m1 = soup.find('table', {'class' :'table_dati','summary':'Company Information'}).find_all('td',{'class' :'name'})
        sector = re.findall(r">(.+?)<",str(m1[5]))[0]
        subsector = re.findall(r">(.+?)<",str(m1[7]))[0]
        market_cap = re.findall(r">(.+?)<",str(m1[9]))[0]
    except:
        sector,subsector,market_cap='n/a','n/a','n/a'
    if not trs:
        df['13'].append([name]+['n/a' for i in range(51)]+[sector,subsector,market_cap])
        df['14'].append([name]+['n/a' for i in range(51)]+[sector,subsector,market_cap])
        df['15'].append([name]+['n/a' for i in range(51)]+[sector,subsector,market_cap])
        df['16'].append([name]+['n/a' for i in range(51)]+[sector,subsector,market_cap])
        df['17'].append([name]+['n/a' for i in range(51)]+[sector,subsector,market_cap])
        return
    cnt = 0
    year_name = []
    for tr in trs:
        ui = []
        for td in tr:
            ui.append(str(td.string).replace('\n',''))
        years = (len(ui)-3)//2
        #print(years)
        if years > 1:
            if len(ui)>5 and ui[3] != 'None' and ui[1] != '\xa0' and ui[1] != 'None':
                data_tmp = ui[3:-1:2]
                cnt += 1
                if cnt == 44:
                    year_name = [data_tmp[i].split('-')[2].replace(' ','') for i in range(len(data_tmp))]
                    #print(year_name[0])
                if cnt > 51:
                    break
                for i in range(years):
                    tmp[i].append(data_tmp[i])
        else:
            if len(ui)==5 and ui[3] != 'None' and ui[1] != '\xa0' and ui[1] != 'None':
                cnt += 1
                if cnt > 51:
                    break
                tmp[0].append(ui[3]) 
    for i in range(len(year_name)):
        tmp[i]+=[sector,subsector,market_cap]
    for i in range(len(year_name)):
        try:
            df[year_name[i]].append(tmp[i])
        except:
            pass
    time.sleep(1)


# chart_data(html,df)
# # pd.DataFrame(df_17,columns = columns)
# #pd.DataFrame(df[13])
# len(df[14])
```
